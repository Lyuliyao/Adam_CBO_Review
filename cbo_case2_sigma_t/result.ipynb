```python
import numpy as np
import matplotlib.pyplot as plt
import jax
jax.config.update("jax_platform_name", "cpu")
```

```python
value  = np.load("result_2/adjust_value.npy")
value = value[:-10]
adjust_mu = value.mean(axis=1)
adjust_sigma = value.std(axis=1)
t = np.arange(adjust_mu.shape[0])
fig, ax = plt.subplots(1)
ax.plot(t, adjust_mu, lw=2, label='mean population 1', color='blue')
ax.fill_between(t, adjust_mu+adjust_sigma, adjust_mu-adjust_sigma, facecolor='blue', alpha=0.5)
```
